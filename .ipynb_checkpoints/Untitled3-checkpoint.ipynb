```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.display.max_columns = 200
sns.set_style('whitegrid')
import warnings
warnings.filterwarnings("ignore")
```

```python
arl = pd.read_csv('/kaggle/input/flight-delays/airlines.csv')
arl.columns = arl.columns.str.lower()
```

```python
df = pd.read_csv('/kaggle/input/flight-delays/flights.csv')
df.columns = df.columns.str.lower()
```

```python
df['std_date'] = pd.to_datetime(df[['year', 'month', 'day']])
df['dep_otp15'] = np.where(df['departure_delay'] <= 15, 1, 0)
```

```python
df.columns
```

```python
g = df.groupby([df['std_date'].dt.weekofyear, 'airline'])['flight_number'].count()
g = g.reset_index()
sns.lineplot(data=g, x='std_date', y='flight_number'
             , hue='airline', palette='Set1');
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left');
```

```python
[x for x in df.columns if 'delay' in x]
```

```python
g = df.query('cancelled == 0 and diverted == 0')
g = g.pivot_table(index=['airline']
              , columns=[df['std_date'].dt.strftime('%Y%m')]
              , values='dep_otp15'
              , aggfunc='mean'
             ) * 100
sns.heatmap(g, annot=True, fmt='0.0f', cmap='RdBu')
```
